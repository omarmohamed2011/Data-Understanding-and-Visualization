# Seasonal Births dataset

## Description: 
Data on number of births every month for approx 135 counties.

## Column Description
Country or area: The name of the country.

Year: the year for which the record is stored.

Month: The name of the month.

Number of births: The total number of births that happed in the month.

```python
## Now I do think of studying  - not each of these columns as an example but only Sweden could be a good example

df_sweden = df[df['Country or Area']=='Sweden']
df_sweden.head(10)
```

```python
df_sweden = df_sweden.drop(['Country or Area'],axis=1)
df_sweden
```

```python
## Let's take in the summation of all months only as the example to understand the data
df_sweden = df_sweden.set_index(['Month'])

df_sweden = df_sweden.loc['Total']
```

```python
df_sweden.head(10)
```

```python
## Now, let's sort the number of births 

df_sweden.sort_values(by='Number_of_Births')
```

```python
## Let's plot the number of births

df_sweden.plot(x= 'Year',y='Number_of_Births')
```

```python
## That's a lot of fluctuations in data especially between 80 and 2000,
## let's print only the above 110 thousands using the query

df_sweden.query('Number_of_Births > 110000')
```

```python
## We can also create a column that says wheter the total num of births in the current year 
## has exceeded 110 thousands or not 
df_sweden['Births_exceeded110'] = df_sweden['Number_of_Births']> 110000
```

```python
## See the num of births with the created new column
for i,j in zip(df_sweden['Number_of_Births'],df_sweden['Births_exceeded110']):
  print(i,j)
```

```python
## Let's analyze data more generally not only Sweden country, but 
## also others and find the average number of births gloabally over the years

df_births = df.groupby(['Year']).mean()
df_births = df_births.reset_index()
df_births.head(20)
```

```python
df_sweden_births = df_sweden.reset_index()[['Year','Number_of_Births']]
df_sweden_births.head(20)
```

```python
df_births.plot(x= 'Year',y='Number_of_Births',c='r')
df_sweden_births.plot(x= 'Year',y='Number_of_Births')
```
